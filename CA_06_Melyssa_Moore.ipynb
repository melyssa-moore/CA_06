```python
#importing packages
import numpy as np
import pandas as pd
from sklearn.neighbors import NearestNeighbors
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
```

```python
#reading in the data and looking at the first few rows and all columns 
movies = pd.read_csv('https://github.com/ArinB/CA05-kNN/raw/master/movies_recommendation_data.csv')
movies
movies.head()
movies.info()
```

```python
#using the columns from IMDB rating and on against the Movie name, movie ID is not being used 
x = movies.iloc[:,2:10]
y = movies[['Movie Name']]
x
```

```python
#creating a KNN classifier to fit the data and show the top 5 movies recommened 
nbrs = KNeighborsClassifier(n_neighbors=5)
nbrs.fit(x,y)
top_5 = nbrs.kneighbors([[7.2,1,1,0,0,0,0,1]])
#creating for loop to print out the top 5
for i in top_5[1][0]:
    print(movies.loc[i]['Movie Name'])

#making the recommender system for "The Post" values, showing the top movie that is recommended 
predicted_values= nbrs.predict([[7.2,1,1,0,0,0,0,1]])
print(predicted_values)
```
